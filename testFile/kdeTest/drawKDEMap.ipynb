```python
%matplotlib inline

# 如果不添加这句，是无法直接在jupyter里看到图的

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sns.set(style="ticks")

# 有一套的参数可以控制绘图元素的比例。
# 首先，让我们通过set()重置默认的参数：
# set( )通过设置参数可以用来设置背景，调色板等，更加常用。
# 有五种seaborn的风格，它们分别是：darkgrid, whitegrid, dark, white, ticks。它们各自适合不同的应用和个人喜好。默认的主题是darkgrid。

df = sns.load_dataset("anscombe")
df.head()

sns.set(style="dark")
rs = np.random.RandomState(50)

f, axes = plt.subplots(3, 3, figsize=(9, 9), sharex=True, sharey=True)

# 设置matplotlib图

for ax, s in zip(axes.flat, np.linspace(0, 3, 10)):

# 绕着cubehelix hue circle相圆旋转起始点。

    cmap = sns.cubehelix_palette(start=s, light=1, as_cmap=True)

    # 创建一个与kdeplot一起使用的cubehelix colormap

    x, y = rs.randn(2, 50)
    print('x', x)
    print('y', y)
    sns.kdeplot(x, y, cmap=cmap, shade=True, cut=5, ax=ax)
    ax.set(xlim=(-3, 3), ylim=(-3, 3))

    # 生成和绘制一个随机的双变量数据集

f.tight_layout()
```
